```python
import pandas as pd
import numpy as np
import lxml
```

```python
tables = pd.read_html('https://en.wikipedia.org/wiki/List_of_cities_in_India_by_population')
```

```python
len(tables) 
```

```python
table1 = tables[0]
print(type(table1))
```

```python
table1.head()
```

```python
table1
```

```python
table1.rename(columns={"Population(2011)[3]":"Population(2011)"},inplace=True)
table1
```

```python
values = ['â€•','â€“','â€”']
table1.replace(values,np.nan,inplace = True)
table1.dropna(axis = 0,inplace = True)
table1["Population(2001)"].unique()
```

```python
table1.iloc[40,2]= 1050721 
table1["Population(2011)"] = table1["Population(2011)"].astype(float)
table1["Population(2001)"] = table1["Population(2001)"].astype(float)
x = ((table1["Population(2011)"]-table1["Population(2001)"])/table1["Population(2001)"])*100
y = (((table1["Population(2011)"]/table1["Population(2001)"])**(1/10))-1)*100
```

```python
table1["basic growth rate %"] = x
table1["average growth rate %"]= y
```

```python
table1
```

```python
data_table1 = table1.sort_values(by=["basic growth rate %","City"],ascending=False)
data_table1.reset_index(drop=True,inplace=True)
data_table1
```

```python
data_table2 = table1.sort_values(by=["average growth rate %"],ascending=False)
data_table2.reset_index(drop=True,inplace=True) 
data_table2
```

```python
rank1=data_table1.sort_values(by="Rank")
```

```python
rank2=data_table2.sort_values(by="Rank")
```
